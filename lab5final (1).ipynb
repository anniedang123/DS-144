```python
from sklearn.preprocessing import LabelEncoder


lb_make = LabelEncoder()
def feature_eng(df3): 
  df = df3.copy() 

  df['Fare'] = df['Fare'].fillna(df['Fare'].median())
  df['Num Relatives'] = df['SibSp'] + df['Parch']
  df['fpp'] = df['Fare']/(df['Num Relatives']+1)
  df['fpp'] = df['fpp'].astype(int)
  df['Cabin'] = df['Cabin'].fillna("U0")
  df['Deck'] = df['Cabin'].map(lambda x: re.compile("([a-zA-Z]+)").search(x).group())
  df['Deck'] = df['Deck'].replace('A', 1)
  df['Deck'] = df['Deck'].replace('B', 2)
  df['Deck'] = df['Deck'].replace('C', 3)
  df['Deck'] = df['Deck'].replace('D', 4)
  df['Deck'] = df['Deck'].replace('E', 5)
  df['Deck'] = df['Deck'].replace('F', 6)
  df['Deck'] = df['Deck'].replace('G', 7)
  df['Deck'] = df['Deck'].replace('U', 8)
  df['Deck'] = df['Deck'].replace('T', 0)
  df['Deck'] = df['Deck'].fillna(0)
  #df['Deck'] = df['Deck'].astype(int)
  df['Title'] = df.Name.str.extract(r'(Mrs|Mr|Miss|Ms|Dr|Rev|Col|Master)', expand=False)
  df['Title'] = df['Title'].replace('Ms', "Miss")
  df['Title'] = df['Title'].replace(['Sir', 'Rev','Capt', 'Dona','Don', 'Dr','Major', 'Countess', 'Lady', 'Jonkheer', 'Col'], 'Rare')
  df['Title'] = df['Title'].replace('Mlle', 'Miss')
  df['Title'] = df['Title'].replace('Mme', 'Mrs')
  df['Title'] = df['Title'].fillna(0)
  df['Title'] = df['Title'].replace('Mr', 1)
  df['Title'] = df['Title'].replace('Miss', 2)
  df['Title'] = df['Title'].replace('Mrs', 3)
  df['Title'] = df['Title'].replace('Master', 4)
  df['Title'] = df['Title'].replace('Rare', 5)


  
  #df['Title2'] = df['Title']
  #df = pd.get_dummies(df, columns=['Title'])
  #df = pd.get_dummies(df, columns=['Embarked'])
  #df['Last_Name'] = df['Name'].apply(lambda x: str.split(x, ",")[0])


  df['Sex'] = df['Sex'].replace('male', 0).replace('female', 1)

  df['Embarked'] = df['Embarked'].fillna("S")
  df['Embarked'] = df['Embarked'].replace('S', 0)
  df['Embarked'] = df['Embarked'].replace('C', 1)
  df['Embarked'] = df['Embarked'].replace('Q', 2)

  df['Age'] = df['Age'].fillna(df['Age'].median())
  df['Age'] = df['Age'].astype(int)
  df.loc[ df['Age'] <= 11, 'Age'] = 0
  df.loc[(df['Age'] > 11) & (df['Age'] <= 18), 'Age'] = 1
  df.loc[(df['Age'] > 18) & (df['Age'] <= 22), 'Age'] = 2
  df.loc[(df['Age'] > 22) & (df['Age'] <= 27), 'Age'] = 3
  df.loc[(df['Age'] > 27) & (df['Age'] <= 33), 'Age'] = 4
  df.loc[(df['Age'] > 33) & (df['Age'] <= 40), 'Age'] = 5
  df.loc[(df['Age'] > 40) & (df['Age'] <= 66), 'Age'] = 6
  df.loc[ df['Age'] > 66, 'Age'] = 6
  df.loc[ df['Fare'] <= 7.91, 'Fare'] = 0
  df.loc[(df['Fare'] > 7.91) & (df['Fare'] <= 14.454), 'Fare'] = 1
  df.loc[(df['Fare'] > 14.454) & (df['Fare'] <= 31), 'Fare']   = 2
  df.loc[(df['Fare'] > 31) & (df['Fare'] <= 99), 'Fare']   = 3
  df.loc[(df['Fare'] > 99) & (df['Fare'] <= 250), 'Fare']   = 4
  df.loc[ df['Fare'] > 250, 'Fare'] = 5
  df['Fare'] = df['Fare'].astype(int)
  df = df.drop(["Ticket", "Cabin", 'Ticket', "PassengerId", "Name", 'Cabin'], axis = 1)
  return df
train = feature_eng(X_train)
test = feature_eng(X_test)
```

```python
from sklearn.model_selection import KFold

def compute_CV_error(model, X_train, Y_train):
    '''
    Split the training data into 4 subsets.
    For each subset, 
        fit a model holding out that subset
        compute the MSE on that subset (the validation set)
    You should be fitting 4 models total.
    Return the average MSE of these 4 folds.

    Args:
        model: an sklearn model with fit and predict functions 
        X_train (data_frame): Training data
        Y_train (data_frame): Label 

    Return:
        the average validation MSE for the 4 splits.
    '''
    kf = KFold(n_splits=4)
    validation_errors = []
    
    for train_idx, valid_idx in kf.split(X_train):
        # split the data
        split_X_train, split_X_valid = X_train.iloc[train_idx], X_train.iloc[valid_idx]
        split_Y_train, split_Y_valid = Y_train.iloc[train_idx], Y_train.iloc[valid_idx]

        # Fit the model on the training split
        model.fit(split_X_train, split_Y_train)
        
        # Compute the RMSE on the validation split
        accuracy = model.score(X_train, Y_train)


        validation_errors.append(accuracy)
        
    return validation_errors

```
